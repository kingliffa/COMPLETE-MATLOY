```python
def cust_age(year):
    age = 2023 - year 
    return age
Ages = cust_age (mkd["Year_Birth"])
print(Ages)
```

```python
mkd["cust_age"] = Ages
```

```python
mkd["cust_age"].unique()
```

```python
a = [mkd["cust_age"].min(), mkd["cust_age"].max()]
print(a)
```

```python
def age_band(age):
    if age >= 25 and age <= 35:
        ageband = "25-35"
    elif  age > 35 and age <= 50:
        ageband = "35-50"
    else:
        ageband = "above 50"
    return ageband
        
    
```

```python
mkd["age_band"] = mkd["cust_age"].apply(age_band)
```

```python
mkd["Kids"] = mkd[["Kidhome", "Teenhome"]].sum(axis = 1)
mkd.drop(["Kidhome", "Teenhome"], axis = 1, inplace = True)
```

```python
mkd["tolt_amnt_good_sld"] = mkd[["MntWines","MntFruits","MntFishProducts","MntMeatProducts","MntSweetProducts","MntGoldProds"]].sum(axis = 1)
```

```python
mkd.head()
```

```python
vol_goods = mkd.groupby('Marital_Status').sum()['tolt_amnt_good_sld']
print(vol_goods)
```

```python
vol_goods.plot(kind = 'bar',)
plt.title('the distribution of goods purchased according to their marital status')
plt.xlabel('marriage_status')
plt.ylabel('volume_goods')
```

```python
av_distro = mkd.groupby('Marital_Status').mean()['Income'].sort_values(ascending = False)
print(av_distro)
```

```python
av_distro.plot(kind = 'bar')
plt.title('the average distribution of income according to their marital status')
```

```python
top_spenders = mkd.groupby('age_band').sum()['tolt_amnt_good_sld']
print(top_spenders)
```

```python
P_explode = [0, 0.3, 0.3]
top_spenders.plot(kind = 'pie', explode = P_explode, shadow = True)mkd
```

```python
mkd.groupby('age_band').mean()['Income']
```

```python
average_income = mkd.pivot_table(index = 'age_band', columns = 'Country', values = 'Income')
print(average_income)
```

```python
average_income.plot(kind = 'barh', stacked=True, color = ['red', 'orange','yellow','green','blue','indigo','violet','black'])
plt.title('the average income of each band according to their countries')
plt.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1, 0.5))
```

```python
mkd.pivot_table(index = 'Kids', columns = 'Country', values = 'tolt_amnt_good_sld', aggfunc = 'sum')
```

```python
av_purchase = mkd.pivot_table(index = 'age_band', columns = 'Country', values = 'tolt_amnt_good_sld')
print(av_purchase)
```

```python
av_purchase.plot(kind = 'barh', stacked=True, color = ['red', 'orange','yellow','green','blue','indigo','violet','black'])
plt.title('the average amount of goods purchased by each age band according to their countries')
plt.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1, 0.5))
```

```python
av_income = mkd.pivot_table(index = 'age_band', columns = 'Education', values = 'Income')
print(av_income)
```

```python
av_income.plot(kind = 'bar', stacked=True, color = ['red', 'orange','yellow','green','blue'])
plt.title('average income of each age band according to their Education status')
plt.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1, 0.5))
```

```python
total_goods = mkd.pivot_table(index = 'age_band', columns = 'Education', values = 'tolt_amnt_good_sld', aggfunc = 'sum')
print(total_goods)
```

```python
total_goods.plot(kind = 'barh', stacked=True, color = ['red', 'orange','yellow','green','blue'])
plt.title('total goods purchased by each age band according to their Education status')
plt.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1, 0.5))
```
